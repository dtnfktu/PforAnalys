## Задача 1

***Создать новый признак Cpu_Company, который будет содержать только название фирмы, которая произвела CPU***

**Решение**

*Возможно в питоне что-то есть для получения первого слова из строки (а здесь в поле CPU первое слово - название производителя). Я делаю свою функцию:*

```python
def get_first_word(first_str):
    tmp = first_str.split(' ')
    return tmp[0] if len(tmp) > 0 else ''
```

*И создаём новый признак в соответствии с заданием:*

```python
df['Cpu_Company'] = df['Cpu'].apply(get_first_word)
```

*Проверяем что у нас в новом признаке*

```python
df['Cpu_Company'].unique()
```

***Создать новый признак Memory_Amount, который будет содержать только количество Gb памяти без указания типа носителя***

**Решение**

*Здесь создаём функцию, возвращающую количество гигабайт типа int*

```python
def get_gigabytes(memory):
    tmp = memory.split(' ')[0]
    ans = 0
    if tmp.endswith('GB'):
        ans = tmp.replace('GB', '')
    elif tmp.endswith('TB'):
        ans = float(tmp.replace('TB', '')) * 1024
    return int(ans)
```

```python
df['Memory_Amount'] = df['Memory'].apply(get_gigabytes)
df.head()
```

***Создать новый признак Memory_Type, который будет содержать только тип носителя (HDD/SDD/др.)***

**Решение**

*Задаём функцию, возвращающую из строки всё, кроме первого слова:*

```python
def get_without_first(word):
    index = word.find(' ')
    return word[index + 1:]
```

*и задаём новый признак:*

```python
df['Memory_Type'] = df['Memory'].apply(get_without_first)
df.head()
```

***Удалите признаки Memory и ScreenResolution***

**Решение**

```python
df.drop(columns = ['Memory','ScreenResolution'], inplace = True)
df.head()
```

***Создайте признак SSD, который изначально равен 0***

**Решение**

```python
df['SSD'] = 0
df.head()
```

***Поставьте в признаке SSD 1, если ноутбук действительно с типом носителя SSD***

***Уберите в признаке Weight значения 'kg' и поменяйте его тип данных на вещественный***

**Решение**

*Здесь чтобы убрать kg по сути надо строку Weight сделать короче на 2 символа*

```python
def cut_kg(weight_str:str):
    return weight_str[:-2] if 'kg' in weight_str else weight_str

df['Weight'] = df['Weight'].apply(cut_kg)
```

```python
df['Weight'] = df['Weight'].astype(float)
df.head()
```

## Задача 4

*Составьте несколько сводных таблиц*

***Найдите среднюю стоимость ноутбуков в зависимости от компании производителя используя метод groupby***

**Решение**

```python
df.groupby('Company').agg({'Price_euros': 'mean'})
```

***Найдите минимальную, среднюю и максимальную стоимости ноутбуков в зависимости от производителя процессора  используя метод groupby***

**Решение**

```python
df.groupby('Cpu_Company').agg({'Price_euros': ['min', 'mean', 'max']})
```

***Постройте таблицу с подсчетом количества ноутбуков в данных в зависимости от производителя CPU и ОЗУ используя метод pivot_table***

**Решение**

```python
df.pivot_table(index='Cpu_Company', columns='Ram', values='laptop_id', aggfunc='count',fill_value=0)
```

***Постройте таблицу с подсчетом средней стоимости ноутбуков в данных в зависимости от операционной системы и GB памяти используя метод pivot_table***

**Решение**

```python
df.groupby(['OpSys', 'Memory_Amount']).agg({'Price_euros': 'mean'}).unstack('Memory_Amount').fillna(0)
```
